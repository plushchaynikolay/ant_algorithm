```python
import numpy as np
L = np.random.randint(1, 50, size=(10, 10))
np.fill_diagonal(L, 0)
print(L)
```

```python
def brute_force(M, inds, dist):
    lvl = len(inds)
    d = []
    for i in range(M.shape[1]):
        if i not in inds and i!=lvl:
            if lvl == (len(M)-1):
                return [dist+M[lvl, i]]
            d = d + brute_force(M, inds+[i], dist+M[lvl, i])
    return d
```
